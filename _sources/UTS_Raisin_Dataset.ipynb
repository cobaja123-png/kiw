2. Menghitung jarak dan menentukan K-NEIGHBOUR

Langkah berikutnya adalah menghitung jarak antara setiap pasangan titik data menggunakan metode jarak Manhattan, yang sering digunakan dalam analisis data untuk mengukur jarak antara titik-titik dalam ruang fitur. Jarak Manhattan dihitung sebagai jumlah dari selisih absolut antara nilai-nilai fitur dari dua titik data. Jarak Manhattan antara dua titik data ð‘¥ dan ð‘¦ dihitung menggunakan rumus berikut:

$$d(x,y) = \sum_{i=1}^{n}(|x_i - y_i|)$$

3. Menghitung  REACHABILITY DENSITY (RD) dan LOCAL REACHABILITY DENSITY(LRD)

Setelah menentukan k-neighbors untuk setiap titik data berdasarkan jarak Manhattan, langkah berikutnya adalah menghitung:

- Reachability Distance (RD): RD antara titik data ð‘¥ dan ð‘¦ adalah maksimum antara k-distance dari titik ð‘¦ dan jarak Manhattan antara ð‘¥ dan ð‘¦.

$$[
\text{reachdist}_k(x,y) = \max(\text{k-distance}(y), \text{dist}(x,y))]
$$

- Local Reachability Density (LRD): LRD untuk titik data ð‘¥ dihitung sebagai kebalikan dari rata-rata dari RD dari k-neighbors terdekatnya.

$$[
\text{lrd}_k(x) = \frac{|N_k(x)|}{\sum_{y \in N_k(x)} \text{reachdist}_k(x,y)}
$$

4. Menghitung  LOCAL OUTLIER FACTOR (LOF)

Langkah terakhir adalah menghitung LOF untuk setiap titik data menggunakan nilai LRD yang telah dihitung sebelumnya. LOF untuk titik data 
ð‘¥ adalah rata-rata dari rasio LRD dari titik ð‘¥ dan LRD dari k-neighbors terdekatnya:

$$
\text{lof}_k(x) = \frac{\sum_{y \in N_k(x)} \text{lrd}_k(y)}{|N_k(x)| \times \text{lrd}_k(x)}
$$

Setelah menghitung LOF untuk setiap titik data, Anda dapat menyusun peringkat dari nilai LOF tersebut. Titik data dengan nilai LOF tertinggi dianggap sebagai outlier atau anomali, karena mereka memiliki nilai LOF yang tinggi dibandingkan dengan tetangga-tetangga mereka. Berikut adalah hasilnya

### PERHITUNGAN  LOCAL OUTLIER FACTOR (LOF) MENGGUNAKAN PYTHON

Untuk perhitungan LOF menggunakan Python, kita bisa menggunakan library scikit-learn. Ini mempermudah dan mempercepat proses deteksi outliers. Berikut adalah kode untuk menghitung LOF:

```
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import numpy as np

# Baca dataset
df = pd.read_csv('/work/Raisin_Dataset.csv')

# Siapkan data untuk LOF (drop kolom 'Class' karena fitur kategorikal)
x = df.drop('Class', axis=1)

# Inisialisasi model LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.040)

# Terapkan LOF pada dataset
y_pred = lof.fit_predict(x) # fit_predict(x): Menerapkan model LOF pada dataset x dan mengembalikan prediksi. Hasil prediksi berupa label -1 untuk outlier dan 1 untuk inlier.
# Skor anomali (semakin rendah, semakin besar kemungkinan outlier)
anomaly_scores = lof.negative_outlier_factor_ # negative_outlier_factor_: Menghasilkan skor anomali untuk setiap sampel dalam dataset. Semakin rendah nilainya, semakin besar kemungkinan sampel tersebut adalah outlier.
# Tambahkan kolom hasil prediksi dan skor anomali ke dalam dataframe
df['LOF_Prediksi'] = y_pred # Menambahkan kolom LOF_Prediksi ke DataFrame df yang berisi hasil prediksi LOF (1 untuk inlier, -1 untuk outlier).
df['LOF_Skor_Anomali'] = anomaly_scores # Menambahkan kolom LOF_Skor_Anomali yang berisi skor anomali dari LOF.

# Visualisasi hasil (opsional)
plt.figure(figsize=(10, 6)) # Membuat figure baru dengan ukuran 10x6.
plt.title("LOF Anomaly Scores") # Memberi judul pada plot.
plt.scatter(np.arange(len(anomaly_scores)), anomaly_scores, c=(y_pred == -1), cmap='coolwarm', edgecolor='k') # Membuat scatter plot dari skor anomali.
plt.axhline(y=-1.3, color='r', linestyle='--')  # Menambahkan garis horizontal pada y = -1.3 untuk memisahkan outlier (opsional, dapat disesuaikan dengan konteks data).
plt.xlabel("Indeks")
plt.ylabel("Skor Anomali LOF")
plt.show()
```

Setelah memahami data, langkah berikutnya adalah melakukan preprocessing data untuk mempersiapkannya sebelum proses pengolahan lebih lanjut. Dalam tahapan preprocessing ini, kami melakukan 3 langkah utama:

Outlier diidentifikasi dengan kondisi di mana nilai 'LOF_Prediksi' tidak sama dengan -1. Dan Berikut adalah langkah-langkahnya:

1. Kami memeriksa setiap baris dalam kolom 'LOF_Prediksi' dari DataFrame df.

2. Baris-baris yang memiliki nilai tidak sama dengan -1 dianggap sebagai outlier dan akan dihapus.

3. Pada data frame baru, df_cleaned dibuat hanya dengan baris-baris yang tidak dianggap sebagai outlier

# ENSEMBLE LEARNING

Setelah dilakukan Pemodelan, ada banyak cara untuk meningkatkan akurasi prediksi dari model yang telah kita buat salah satunya adalah **Ensemble Learning**. 

Apa itu Ensemble Learning?

## PENGERTIAN ENSEMBLE LEARNING

&nbsp;&nbsp;&nbsp;&nbsp;Ensemble learning adalah teknik yang digunakan untuk meningkatkan kinerja dan akurasi model-model pembelajaran mesin dengan menggabungkan hasil prediksi dari beberapa model atau algoritma yang berbeda. 

&nbsp;&nbsp;&nbsp;&nbsp;Istilah ensemble merujuk pada konsep menggabungkan banyak elemen menjadi satu kesatuan. Konsep penggabungan berbagai pendekatan pembelajaran mesin diharapkan dapat mengatasi kelemahan individu dari masing-masing model dan menciptakan prediksi yang lebih akurat dan stabil.

&nbsp;&nbsp;&nbsp;&nbsp;Cara kerja dari metode ensemble melibatkan penggunaan beberapa model yang beragam dalam satu tim. Hasil prediksi dari masing-masing model tersebut digabungkan untuk membuat keputusan akhir.

&nbsp;&nbsp;&nbsp;&nbsp;Pada project kali ini kita akan menggunakan 2 metode dalam ensemble learning yaitu **Stacking** dan **Bagging** dan mencari metode mana yang terbaik untuk data kita. 

## METODE STACKING

### 1. Pengertian Metode Stacking

&nbsp;&nbsp;&nbsp;&nbsp;Stacking adalah sebuah metode dimana sebuah model ditraining untuk menggabungkan beberapa individual model yang disebut *first-level learner*, sedangkan yang digabungkan disebut *second-level learner* atau *meta-learner*. 

&nbsp;&nbsp;&nbsp;&nbsp;Stacking melibatkan dua atau lebih *base learner* atau model dasar sebagai model level-1, dan meta-learner yang menggabungkan prediksi base learner disebut sebagai level-2. Model yang berbeda digunakan oleh base learner untuk belajar dari suatu dataset. Output dari masing-masing model base learner dikumpulkan untuk membuat dataset baru. Di dalam dataset yang baru, setiap sampel berhubungan dengan nilai pada dataset awal yang seharusnya diprediksi. Selanjutnya dataset tersebut digunakan oleh model meta-learner hingga mendapatkan hasil akhir.
